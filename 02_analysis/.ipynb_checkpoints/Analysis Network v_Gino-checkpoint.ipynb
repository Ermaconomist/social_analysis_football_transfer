#### Plotting Networt to show which Teams deliver Players to which language in the Season 20/21
* Saving the plot as a .pdf-File (fig.savefig('network20.pdf'))

```python
fig = plt.figure(figsize=(10, 10)) 
g = nx.from_pandas_edgelist(df_temp20, 
                            source = 'ToLanguage',
                            target = 'From',
                            edge_attr = 'To',
                            create_using = nx.Graph())
nx.draw(g, with_labels = True) 
plt.axis('equal') 
plt.show()
fig.savefig('network20.pdf')
```

```python
fig = plt.figure(figsize=(10, 10)) 
g = nx.from_pandas_edgelist(df_temp20, 
                            source = 'FromLanguage',
                            target = 'ToLanguage',
                            edge_attr = 'ToLanguageFamily',
                            create_using = nx.MultiDiGraph())
nx.draw(g, with_labels = True) 
plt.axis('equal') 
plt.show()
fig.savefig('networkTest20.pdf')
```

```python
pos = nx.spring_layout(g)

betCent = nx.betweenness_centrality(g, normalized=True, endpoints=True)
node_color = [20000.0 * g.degree(v) for v in g]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize = (20,20))
nx.draw_networkx(g, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off')
```
